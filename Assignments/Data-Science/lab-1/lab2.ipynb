```python
import pandas as pd
```

```python
data={
    'S.N':[1,2,3,4,5],
    'Name':['Samip','Sanskar','Sumit','Samyog','Pratik'],
    'Address':['Brt','Bkt','Lpr','Dhd','Btl'],
    'Age':[20,26,22,23,24]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.to_csv('FOD.csv',index=False)
```

```python
df2=pd.read_csv('FOD.csv')
df2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df2['Age'],df2['S.N'],color="#DD129996",marker='*',linestyle='--',linewidth=2)
plt.xlabel('Age')
plt.ylabel('S.N')
plt.title('Age VS S.N')
plt.yticks(df2['S.N'])
plt.grid()
plt.show()
```

```python
plt.bar(df2['Age'],df2['S.N'],color="#2A205799",width=0.9,edgecolor="#F0080894",hatch='*')
plt.xlabel('Age')
plt.ylabel('S.N')
plt.title('Age VS S.N')
plt.grid()
plt.show()
```

```python
plt.barh(df2['Age'],df2['S.N'],color="#2EC40095",height=0.9,edgecolor="#F0080894",hatch='o')
plt.xlabel('Age')
plt.ylabel('S.N')
plt.title('Age VS S.N')
plt.grid()
plt.show()
```

```python
HATCHES=('/','o','*','o','*')
COLOR=('pink','black','green','yellow','purple')
plt.bar(df2['Age'],df2['S.N'],color=COLOR,width=0.9,edgecolor="#F0080894",hatch=HATCHES)
plt.xlabel('Age')
plt.ylabel('S.N')
plt.title('Age VS S.N')
plt.grid()
plt.show()
```

```python
EXPLODE=(0,0.2,0,0,0.2)
plt.pie(df['Age'],labels=df['S.N'],autopct='%1.1f%%',hatch='/',explode=EXPLODE,shadow=True,startangle=25)
plt.show()
```

```python
plt.hist(df['Age'],hatch='*0',bins=11)
plt.show()
```

```python
plt.boxplot(df['Age'])
plt.xticks([1],['Age'])
plt.show()
```

```python
plt.boxplot([df['Age'],df['S.N']])
plt.xticks([1,2],['Age','S.N'])
plt.show()
```
